```python
import sys
import os
import nbimporter
from backend_for_querying import BackendForQuerying

# Add the parent directory to the sys.path (so it can find the backend file)
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from backend_for_querying import BackendForQuerying
from flask import Flask, request, jsonify

app = Flask(__name__)

# Initialize the backend querying system
backend = BackendForQuerying()

@app.route('/query', methods=['POST'])
def query():
    # Get the query from the frontend (in JSON format)
    data = request.get_json()
    query_text = data.get('query', '')
    
    # Get the top results from the backend
    results = backend.query(query_text, n_results=3)
    
    # Return the results as a JSON response
    return jsonify(results)

if __name__ == '__main__':
    app.run(debug=True)
```
